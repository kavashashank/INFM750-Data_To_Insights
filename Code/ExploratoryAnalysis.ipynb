```python
corr = stats.pearsonr( df['fb_fan_count'],df['NormalizedAnnualVisCnt'])
print("FB Likes Count vs Normalized Vistor Count : " + str(corr))

corr = stats.pearsonr( df['fb_talking_about_count'],df['NormalizedAnnualVisCnt'])
print("FB Talking about Count vs Normalized Vistor Count : " + str(corr))

corr = stats.pearsonr( df['fb_checkins'],df['NormalizedAnnualVisCnt'])
print("FB Checkins Count vs Normalized Vistor: " + str(corr))

print("\nMutliple Regression Facebook")

X = df[['fb_fan_count','fb_talking_about_count','fb_checkins']]
y = df['AverageVisitors']

model = sm.OLS(y, X).fit()
predictions = model.predict(X)
model.summary()
```

```python
#Graph
print("Distribution of FB Likes Count")
df1 = df.sort_values(['fb_fan_count'])
h1 = df1['fb_fan_count']
fit = stats.norm.pdf(h1, np.mean(h1), np.std(h1))  #this is a fitting indeed
pl.plot(h1,fit,'-o')
#pl.hist(h1,normed=True)      #use this to draw histogram of your data
pl.show()

print("Distribution of FB Talking About Count")
df1 = df.sort_values(['fb_talking_about_count'])
h1 = df1['fb_talking_about_count']
fit = stats.norm.pdf(h1, np.mean(h1), np.std(h1))  #this is a fitting indeed
pl.plot(h1,fit,'-o')
#pl.hist(h1,normed=True)      #use this to draw histogram of your data
pl.show()

print("Distribution of FB Checkins Count")
df1 = df.sort_values(['fb_checkins'])
h1 = df1['fb_checkins']
fit = stats.norm.pdf(h1, np.mean(h1), np.std(h1))  #this is a fitting indeed
pl.plot(h1,fit,'-o')
#pl.hist(h1,normed=True)      #use this to draw histogram of your data
pl.show()
```
